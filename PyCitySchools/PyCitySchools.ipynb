```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
summary_df = pd.DataFrame({
    "Total Schools":[len(school_data)],
    "Total Students":[len(student_data)],
    "Total Budget":[school_data.budget.sum()],
    "Average Math Score":[round(student_data.math_score.mean(),2)],
    "Average Reading Score":[round(student_data.reading_score.mean(),2)],
    "% Passing Math":[round(len(student_data.loc[student_data['math_score'] >= 70])/len(student_data)*100,2)],
    "% Passing Reading":[round(len(student_data.loc[student_data['reading_score'] >= 70])/len(student_data)*100,2)],
    "Overall Passing Rate":[round(100*(len(student_data.loc[student_data['reading_score'] >= 70])/len(student_data)\
                                   +len(student_data.loc[student_data['math_score'] >= 70])/len(student_data))/2,2)]
})
summary_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

math_avgs = round(school_data_complete.groupby('School ID')['math_score'].mean(),2)
reading_avgs = round(school_data_complete.groupby('School ID')['reading_score'].mean(),2)

passing_math = school_data_complete.loc[school_data_complete['math_score'] >= 70]
passing_math = passing_math.groupby('School ID')['math_score'].count()
perc_passing_math = round(100*passing_math / school_data['size'],2)

passing_reading = school_data_complete.loc[school_data_complete['reading_score'] >= 70]
passing_reading = passing_reading.groupby('School ID')['math_score'].count()
perc_passing_reading = round(100*passing_reading / school_data['size'],2)

passing_rate = round((perc_passing_math + perc_passing_reading) / 2,2)

school_data.reset_index(inplace=True)
school_data.set_index('School ID', inplace=True)

names=school_data.school_name
types = school_data.type
sizes = school_data['size']
budgets = school_data.budget
bps = round(budgets/sizes,2)

schools_df = pd.DataFrame({
    "School Name":names,
    "School Type":types,
    "Total Students":sizes,
    "Total Budget":budgets,
    "Budget per Student":bps,
    "Avg Math Score":math_avgs,
    "Avg Reading Score":reading_avgs,
    "% Passing Math":perc_passing_math,
    "% Passing Reading":perc_passing_reading,
    "Overall Passing Rate":passing_rate
})
schools_df
```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
top_schools = schools_df.sort_values('Overall Passing Rate',ascending=False).head()
top_schools
```

## Bottom Performing Schools (By Passing Rate)

* Sort and display the five worst-performing schools

```python
bottom_schools = schools_df.sort_values('Overall Passing Rate',ascending=False).tail()
bottom_schools
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", ">$645"]
schools_df['Spending Group'] = pd.cut(schools_df['Budget per Student'], spending_bins, labels=group_names)
schools_df['Passing Math'] = passing_math
schools_df['Passing Reading'] = passing_reading
spending_maths = round(schools_df.groupby('Spending Group')['Avg Math Score'].mean(),2)
spending_reads = round(schools_df.groupby('Spending Group')['Avg Reading Score'].mean(),2)
#spending_perc_pass_math = round(schools_df.groupby('Spending Group')['% Passing Math'].mean(),2) #avg of %'s
spending_perc_pass_math = round(schools_df.groupby('Spending Group')['Passing Math'].sum()\
                                / schools_df.groupby('Spending Group')['Total Students'].sum()*100,2) 
spending_perc_pass_reading = round(schools_df.groupby('Spending Group')['Passing Reading'].sum()\
                                  / schools_df.groupby('Spending Group')['Total Students'].sum()*100,2)
spending_overall_pass = round((spending_perc_pass_math + spending_perc_pass_reading) / 2,2)

spending = pd.DataFrame({
    "Average Math Score":spending_maths,
    "Average Reading Score":spending_reads,
    "% Passing Math":spending_perc_pass_math,
    "% Passing Reading":spending_perc_pass_reading,
    "Overall Passing Rate":spending_overall_pass
})
spending
```

## Scores by School Size

* Perform the same operations as above, based on school size.

```python
size_bins = [0, 1000, 2000, 5000]
group_names = ["Small (<1000)", "Medium (1000-2000)", "Large (>2000)"]
schools_df['Size Group'] = pd.cut(schools_df['Total Students'], size_bins, labels=group_names)
size_maths = round(schools_df.groupby('Size Group')['Avg Math Score'].mean(),2)
size_reads = round(schools_df.groupby('Size Group')['Avg Reading Score'].mean(),2)
size_perc_pass_math = round(schools_df.groupby('Size Group')['Passing Math'].sum()\
                           /schools_df.groupby('Size Group')['Total Students'].sum()*100,2)
size_perc_pass_read = round(schools_df.groupby('Size Group')['Passing Reading'].sum()\
                           /schools_df.groupby('Size Group')['Total Students'].sum()*100,2)
size_overall_pass = round((size_perc_pass_math + size_perc_pass_read) /2,2)
size_matters = pd.DataFrame({
    "Average Math Score":size_maths,
    "Average Reading Score":size_reads,
    "% Passing Math":size_perc_pass_math,
    "% Passing Reading":size_perc_pass_read,
    "Overall Passing Rate":size_overall_pass
})
size_matters
```

```python
type_maths = round(schools_df.groupby('School Type')['Avg Math Score'].mean(),2)
type_reads = round(schools_df.groupby('School Type')['Avg Reading Score'].mean(),2)
type_perc_pass_math = round(schools_df.groupby('School Type')['Passing Math'].sum()\
                           /schools_df.groupby('School Type')['Total Students'].sum()*100,2)
type_perc_pass_read = round(schools_df.groupby('School Type')['Passing Reading'].sum()\
                           /schools_df.groupby('School Type')['Total Students'].sum()*100,2)
type_overall_pass = round((type_perc_pass_math + type_perc_pass_read) / 2,2)
types_df = pd.DataFrame({
    "Average Math Score":type_maths,
    "Average Reading Score":type_reads,
    "% Passing Math":type_perc_pass_math,
    "% Passing Reading":type_perc_pass_read,
    "Overall Passing Rate":type_overall_pass
})
types_df
```
